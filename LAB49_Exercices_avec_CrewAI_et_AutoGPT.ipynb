## 2ï¸âƒ£ CrewAIÂ : DÃ©finir une Ã©quipe simpleÂ :

CrewAI vous permet dâ€™attribuer des rÃ´les et de dÃ©lÃ©guer des tÃ¢ches Ã  diffÃ©rents agents travaillant en Ã©quipe.

## ğŸ¯ Concept clÃ©

- **Agents**Â : Membres de lâ€™Ã©quipe ayant des rÃ´les et des objectifs spÃ©cifiques
- **TÃ¢ches**Â : Missions assignÃ©es aux agents
- **Ã‰quipe**Â : Lâ€™orchestrateur qui gÃ¨re le flux de travail


````python
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
import os
````

### ğŸ‘¥ Step 1: DÃ©finition des Agents :

````python
# Define the Researcher agent
researcher = Agent(
    role="Researcher",
    goal="Gather facts about Agentic AI.",
    backstory="Curious and thorough, always finds supporting evidence. "
              "You excel at finding credible sources and synthesizing information.",
    verbose=True,
    allow_delegation=False
)

# Define the Writer agent
writer = Agent(
    role="Writer",
    goal="Write concise blog-style summaries.",
    backstory="Skilled communicator who explains complex topics simply. "
              "You have a talent for making technical concepts accessible.",
    verbose=True,
    allow_delegation=False
)

# Define the Editor agent
editor = Agent(
    role="Editor",
    goal="Review and refine drafts for clarity.",
    backstory="Sharp critic with a knack for improving readability. "
              "You catch errors and improve structure and flow.",
    verbose=True,
    allow_delegation=False
)

print("âœ… Agents created: Researcher, Writer, and Editor")
````

### ğŸ“ Step 2: DÃ©finition des TÃ¢ches (Tasks) :

````python
# Task 1: Research
task1 = Task(
    description="Research key benefits of Agentic AI. Focus on: "
                "autonomy, decision-making capabilities, real-world applications, "
                "and current limitations. Provide at least 3-5 concrete benefits.",
    agent=researcher,
    expected_output="In French : A detailed list of key benefits with explanations"
)

# Task 2: Write
task2 = Task(
    description="Draft a 3-paragraph blog post on Agentic AI based on the research. "
                "Include: an introduction, main benefits, and conclusion. "
                "Keep it engaging and accessible to a general audience.",
    agent=writer,
    expected_output="In French : A well-structured 3-paragraph blog post"
)

# Task 3: Edit
task3 = Task(
    description="Edit the draft for clarity and conciseness. "
                "Fix any grammatical errors, improve flow, "
                "and ensure technical accuracy. Provide the final polished version.",
    agent=editor,
    expected_output="In French : A polished, publication-ready blog post"
)

print("âœ… Tasks created: Research â†’ Write â†’ Edit")
````

### ğŸš€ Step 3: Assemblage et exÃ©cution de l'Ã©quie (Crew) :

````python
# CrÃ©ation du workflow "crew"
crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[task1, task2, task3],
    process=Process.sequential,  # ExÃ©cution des tÃ¢ches en sÃ©quence
    verbose=True
)

# ExÃ©cution du workflow "crew"
print("\n" + "="*60)
print("ğŸ¬ Lancement du workflow CrewAI")
print("="*60 + "\n")

result = crew.kickoff()

print("\n" + "="*60)
print("âœ… CrewAI Workflow : exÃ©cution finalisÃ©e !")
print("="*60)
print("\nğŸ“„ RÃ©sultat Final:\n")
print(result)
````

### ğŸ“Points clÃ©s Ã  retenir :

**CrewAI orchestre les rÃ´les au sein d'un flux de travail structurÃ©**Â :
- Chaque agent se voit attribuer une tÃ¢che spÃ©cifique
- Les agents collaborent de maniÃ¨re sÃ©quentielle ou parallÃ¨le pour atteindre un objectif commun.
- Cette approche est idÃ©ale pour les tÃ¢ches qui tirent profit de la spÃ©cialisation et d'une rÃ©partition claire des tÃ¢ches.

### ğŸ¤– Agent autonome de type AutoGPT :

Cet agent effectuera les opÃ©rations suivantesÂ :

1. Recevoir un objectif

2. Le dÃ©composer en sous-tÃ¢ches

3. ExÃ©cuter chaque sous-tÃ¢che

4. SynthÃ©tiser les rÃ©sultats

````python
import json
from typing import List
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field

class SubtaskList(BaseModel):
    tasks: List[str] = Field(description="List of 3-5 subtasks")

class SimpleAutoGPT:
    def __init__(self, model="gpt-4o-mini", max_iterations=5):
        self.llm = ChatOpenAI(model=model, temperature=0)
        self.max_iterations = max_iterations
        self.iteration_count = 0
        # On utilise un parser explicite au lieu du helper .with_structured_output()
        self.parser = JsonOutputParser(pydantic_object=SubtaskList)

    def decompose_goal(self, goal: str) -> List[str]:
        system_prompt = (
            "You are an AI planner. Break down the goal into 3-5 concrete steps. "
            "Your output must be a JSON object with a 'tasks' key containing a list of strings."
        )
        # On injecte les instructions de formatage du parser
        format_instructions = self.parser.get_format_instructions()
        user_prompt = f"Goal: {goal}\n\n{format_instructions}"

        response = self.llm.invoke([
            SystemMessage(content=system_prompt),
            HumanMessage(content=user_prompt)
        ])

        try:
            # On parse manuellement le contenu textuel
            parsed_output = self.parser.parse(response.content)
            return parsed_output.get("tasks", [])
        except:
            return [f"Recherche sur {goal}"]

    def execute_subtask(self, subtask: str, context: str = "") -> str:
        prompt = f"Execute this subtask: {subtask}\n\nContext: {context}"
        response = self.llm.invoke([HumanMessage(content=prompt)])
        return response.content

    def synthesize_results(self, goal: str, results: List[str]) -> str:
        results_text = "\n".join(results)
        prompt = f"Summarize these findings for the goal: {goal}\n\nFindings:\n{results_text}\n\nLimit to ~150 words."
        response = self.llm.invoke([HumanMessage(content=prompt)])
        return response.content

    def run(self, goal: str):
        print(f"\nğŸ¯ OBJECTIF : {goal}")

        subtasks = self.decompose_goal(goal)
        print(f"ğŸ“‹ Plan Ã©tabli ({len(subtasks)} Ã©tapes)")

        results = []
        context = ""

        for i, task in enumerate(subtasks, 1):
            if self.iteration_count >= self.max_iterations: break
            print(f"  â–¶ï¸ Ã‰tape {i}: {task}")
            res = self.execute_subtask(task, context)
            results.append(res)
            context += f"\n- {res[:100]}..."
            self.iteration_count += 1

        print("ğŸ”„ SynthÃ¨se finale...")
        final_output = self.synthesize_results(goal, results)
        print(f"\nâœ… RÃ‰PONSE FINALE :\n{'-'*30}\n{final_output}")
        return final_output
````

### ğŸš€ ExÃ©cution de l'agent AutoGPT-Style :

````python
import os

# 1. Initialisation de l'agent avec la nouvelle classe
#  10 itÃ©rations max
auto_agent = SimpleAutoGPT(model="gpt-4o-mini", max_iterations=10)

# 2. DÃ©finition de l'objectif
# Note : L'agent peut traiter l'anglais et le franÃ§ais sans problÃ¨me
goal = "Give the results in French : Find 3 real-world applications of multi-agent AI and summarize them in ~150 words."

# 3. Interface de lancement
print("\n" + "ğŸ¤–" * 40)
print(f"{'AutoGPT-Style Agent Demo':^40}")
print("ğŸ¤–" * 40 + "\n")

# 4. ExÃ©cution du workflow (Planification -> ExÃ©cution -> SynthÃ¨se)
try:
    result = auto_agent.run(goal)
except Exception as e:
    print(f"âŒ Une erreur est survenue : {e}")
````

### ğŸ“Points clÃ©s Ã  retenir :

**Les agents de type AutoGPT dÃ©composent itÃ©rativement les objectifs en sous-tÃ¢ches** et les exÃ©cutent de maniÃ¨re autonome.

L'agentÂ :

1. **Planifie**Â : DÃ©compose l'objectif en Ã©tapes gÃ©rables.

2. **Agit**Â : ExÃ©cute chaque Ã©tape indÃ©pendamment.

3. **Analyse**Â : Construit le contexte Ã  partir des Ã©tapes prÃ©cÃ©dentes.

4. **SynthÃ©tise**Â : Combine les rÃ©sultats pour obtenir une rÃ©ponse finale.

Cette approche est plus flexible, mais moins prÃ©visible que le flux de travail structurÃ© de CrewAI.
